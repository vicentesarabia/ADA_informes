Entrada: Se ingresa un arreglo de **n** elementos [1,7,3,2,5,...,n]. 

 
 

Salida: Sale el arreglo de n√∫meros ordenados de menor a mayor  [1,2,3,5,7,....,n]

El principal problema es el ordenamiento de mayor a menor un arreglo de **n** elementos y resolverlo de una forma eficiente, por el cual el algoritmo Insertion Sort puede resolver este problema (ya resuelto por varios algoritmos) 

# 2.- InsertionSort


**2.1 Descripcion del algoritmo InsertionSort**

La entrada de este algoritmo es una secuencia de n√∫meros los cuales ser√°n ordenados de menor a mayor en un arreglo. 

 

1.- En la primera iteraci√≥n se tomar√° la en√©sima posici√≥n defini√©ndola en una **"key"** para as√≠ poder guardar esta posici√≥n y realizar el cambia de posici√≥n si es debido, luego entra en otro ciclo para as√≠ poder realizar las comparaciones. 

 
 

2.- Si la **inesima-1** posici√≥n es mayor a la **"key"** **(array[i-1]>key)**, se intercambian las posiciones el uno con el otro.

 

3.- Si no es la primera iteraci√≥n se sigue comparando las posiciones que se van renovando por cada iteraci√≥n en la que se cumple la condici√≥n **(array[i-1]>key)**, y as√≠ hasta que el arreglo se ordene de menor a mayor.
 

El arreglo se recorre cada vez con una posici√≥n m√°s, ya que parte de una "en√©sima" y se asegura que esta posici√≥n este bien posicionada en la cantidad de posiciones que hay hasta el momento, en si se podr√≠a decir que en el arreglo se van creando sub arreglos, partiendo siempre de izquierda a derecha, la condici√≥n (array[i-1]>key) si va retrocediendo para que el n√∫mero m√°s peque√±o quede a la izquierda de este sub arreglo, para as√≠ ir ordenando estos sub arreglos hasta que el arreglo est√° completamente ordenado.

  **2.2 Implementacion del algoritmo InsertionSort**

Aqu√≠ la implementaci√≥n del algoritmo Insertion Sort con verbose=False con un arreglo de ejemplo

```python

from termcolor import colored

def InsertionSort(array,verbose=False):
  iteraciones=0
  for i in range(1,len(array)):
    key = array[i]
    paso=0
    a=0
    
    
    if verbose==True:
       print("pasada",i)
       salir=0

    for k in range(i-1,-1,-1):
        
        
        if array[k]>key: #condicion que compara la clave con las pocisiones que estan a su izquierda
            iteraciones+=1
            #intercambio de posiciones
            array[k+1]=array[k] 
            array[k]=key
            if verbose==True:
              print(str(array[:k])[1:-1],",",colored(f"{str(array[k:k+2])[1:-1]}","green"),",",str(array[k+2:])[1:-1])
              paso+=1
            else:
              paso+=1
        if verbose==True and paso==0: ##condicion que retorna cuando el arreglo completo ya esta ordenado
          if salir==1 and i==len(array)-1:
            return array,iteraciones

          if salir==0: #muestra el subarreglo cuando ya esta ordenado
            print(str(array)[1:-1])

          salir+=1
        if paso==0:#condicion para salir si esta ordenado ya el subarreglo
          a+=1
          if a==1:
            
            break
        
          

    if key<array[0]:
        array[0]=key  
        
  
  return array,iteraciones

array = [3, 9, 4, 7, 2, 5]
print("Entrada: ",array)
i=0
A, iteraciones = InsertionSort(array)
print("Salida: ",array)
print("Iteraciones: ", iteraciones)
```

# 3.- Tiempo de ejecucion InsertionSort


**3.1 teorema de tiempo de ejecucion**

El tiempo de ejecuci√≥n que tiene el algoritmo InsertionSort es de $O(n^2)$ en el peor caso 


**3.1 Prueba del teorema**

El peor caso del InsertionSort es cuando el arreglo va de mayor a menor, ya que tendr√° que hacer la mayor cantidad de iteraciones posibles al tener que recorrer el arreglo n-1 veces, moviendo siempre el n√∫mero menor al principio de cada subarreglo que se crea 

**Ejemplo**



```python
arreglo2=[6,5,4,3,2,1]
print("arreglo = " ,arreglo2)
arreglo2,iteraciones=InsertionSort(arreglo2,True)
print("Arreglo = ",arreglo2)
print("Cantidad de iteraciones",iteraciones)
```

Se puede ver que cada vez que va aumentando el √≠ndice del subarreglo se tiene que recorrer todo el sub arreglo para ordenar de forma exitosa el arreglo, ya que tendr√° que mover la √∫ltima posici√≥n hasta el principio del arreglo o sub arreglo 

Esto genera una sumatoria ya que por cada pasada la cantidad de iteraciones va aumentando, primero solo 1 luego 2,3,4 y esto aumenta seg√∫n el tama√±o que tenga el arreglo(n-1) por lo que se genera la ecuaci√≥n $n(n-1)*1/2$ la cual corresponde a la complejidad temporal de $O(n^2)$ 


 
# 4.- Correctitud



**4.1 Teorema**


_El algoritmo Insertion Sort genera un arreglo $[a_1',a_2', \ldots ,a_n']$ mediante permutaciones de los mismos elementos de entrada, de tal forma que $a_1' \leq a_2' \leq \ldots \leq a_n'$_


**4.2 Prueba del teorema**


Anteriormente se mostr√≥ que por cada iteraci√≥n $i$, el elemento $a_i$ es comparado con los primeros $i-1$ elementos del arreglo, y en caso de encontrarse un elemento $a_j$ menor, los elementos $[a_{j} \ldots a_{i-1}]$ ser√°n desplazados una posici√≥n a la derecha, y el elemento $a_i$ ser√° insertado donde se encontraba previamente $a_j$.

A partir de esto, probaremos la siguiente **propiedad invariante de bucle** para el algoritmo **Insertion Sort**:

> Al comienzo de cada iteraci√≥n, los primeros $i-1$ elementos que corresponden al arreglo original se encuentran ordenados de menor a mayor.


**Inicializacion**


Al comienzo de la primera iteraci√≥n $(i = 1)$, los primeros $i - 1 = 0$ elementos del arreglo cumplen con el teorema. Esto es trivialmente cierto, puesto que al ser un √∫nico elemento obviamente se encuentra ordenado.


**Mantencion**


Mediante inducci√≥n, podemos asumir que en cada iteraci√≥n se cumple la propiedad invariante de bucle, es decir, el sub-arreglo  [a0,‚Ä¶,ai‚àí1]  est√° ordenado de tal forma que  a0‚â§‚Ä¶‚â§ai‚àí1 

Por cada iteraci√≥n, el sub-arreglo se expande, pero se mantiene ordenado, ya que cuando se inserta un elemento  Œª , se insertar√° luego del primer elemento  ‚â§Œª  y antes del primer elemento  ‚â•Œª , manteniendo el orden de menor a mayor y cumpliendose la propiedad.



**Correctitud**
Ya que se demostr√≥ que la propiedad de bucle invariante se cumple al inicio del bucle y se mantiene en cada iteraci√≥n, podemos decir que al **finalizar la $n$-√©sima iteraci√≥n del algoritmo Insertion Sort**, se genera un arreglo $[a_1',a_2',...,a_n']$, con los elementos iniciales del arreglo de entrada ordenados de menor a mayor. $\Box$
