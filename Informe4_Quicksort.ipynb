#1.- Problema

Entrada: Un arreglo de n elementos [3,76,4,9.....n]

Salida: El arreglo de entrada pero esta vez ordenado de menor a mayor sus elementos [3,4,9,76...n]


arbol-particion.gif

El problema es el ordenamiento de un arreglo, en el cual los elementos de este arreglo quedan ordenados de menor a mayor, según sus elementos.

# 2.-Descripción del algoritmo

**2.1 descripción función partición**

En la función partición lo principal es la elección del pivote, que en nuestro caso podemos elegir que el pivote este a la izquierda, derecha o la mediana del arreglo con elementos al azar, cuando se define un pivote lo que se hace es tener dos puntos que serán los límites de arreglo, inicio(i, izquierda) y final(j, derecha), en la que la variable de la izquierda va en aumento y el de la derecha en decremento.
Una de las condiciones es que todos los elementos que sean menores del pivote estén a la izquierda y todos los elementos mayores al pivote a la derecha.

Los índices de izquierda y derecha se van acercando, solo mientras el arreglo este ordenado, o sea todos los elementos menores a la izquierda del pivote  y todos los mayores a la derecha del pivote

Esta función retorna la posición donde quedo para así el algoritmo quicksort haga sus recursiones en dos mitades.

**2.4 Descripción función quicksort**

La función Quicksort es la función recursiva del algoritmo, en la cual divide el arreglo hasta que este ordenado en dos mitades, en la cual la división se crea por el índice retornado en la función partición, en la cual se recorre recursivamente hasta que este queda con un elemento que está ordenado.


**2.4 Implementación función Quicksort**

A continuación la implementación de la función Quicksort

```python
def determine_median(a, b, c):
        if a > b:
            if a < c:
                median = a
            elif b > c:
                median = b
            else:
                median = c
        else:
            if a > c:
                median = a
            elif b < c:
                median = b
            else:
                median = c
        return median

def quicksort(arreglo, izquierda, derecha,vari,verbose=False):
    if izquierda < derecha:
        if vari==1:
          
          indiceParticion = particion(arreglo, izquierda, derecha,vari,verbose)
          
          quicksort(arreglo, izquierda, indiceParticion,vari,verbose)
          quicksort(arreglo, indiceParticion + 1, derecha,vari,verbose)
          
        if vari==2:
          indiceParticion = particion(arreglo, izquierda, derecha,vari,verbose)
          
          quicksort(arreglo, izquierda, indiceParticion-1,vari,verbose)
          quicksort(arreglo, indiceParticion + 1, derecha,vari,verbose)
        if vari==3:
          indiceParticion = particion(arreglo, izquierda, derecha,vari,verbose)
          quicksort(arreglo, izquierda, indiceParticion-1,vari,verbose)
          quicksort(arreglo, indiceParticion + 1, derecha,vari,verbose)



    return arreglo

"""
Modo de uso:
"""

arreglo = [2, 24, 10, 1, 22, 19]
print("Antes de ordenarlo: ")
print(arreglo)
cont1=0;
vari=1;
quicksort(arreglo, 0, len(arreglo) - 1,vari,verbose=False)
print("Después de ordenarlo pivote inicio del arreglo: ")
print(arreglo)
print("Cantidad de comparaciones pivote inicio del arreglo: "  ,cont1)
cont2=0;
vari=2
arreglo = [2, 24, 10, 1, 22, 19]
quicksort(arreglo, 0, len(arreglo) - 1,vari,verbose=False)
print("Después de ordenarlo pivote al fina del arreglo: ")
print(arreglo)
print("Cantidad de comparaciones pivote final del arreglo: "  ,cont2)
cont3=0;
vari=3
arreglo = [2, 24, 10, 1, 22, 19]
quicksort(arreglo, 0, len(arreglo) - 1,vari,verbose=False)
print("Después de ordenarlo pivote mediana de elementos al azar: ")
print(arreglo)
print("Cantidad de comparaciones pivote mediana de elementos al azar: "  ,cont3)

```

## 3.- Correctitud

**3.1 Teorema correctitud función partición**

Entra un arreglo de $[izq....der]$ con der-izq+1 elementos e indices **izq** y **der*** y retorna el indiceParticion tal que $arreglo[izq]<=arreglo[indiceParticion]<=arreglo[der]$

**3.2 Prueba correctitud función partición**

Para probar nuestro teorema utilizaremos la siguiente propiedad invariante

1.- $izq\leq k\leq i, entonces ,arreglo[k]\leq x$

2.- $i< k< j ,entonces ,arreglo[k]> x$

3.- $k=izq, entonces, arreglo[k]=x$

**Inicialización:**

Al principio de la iteración del algoritmo quicksort el índice i=izq-1 y j=izq,
como no existe el k tal que $p\leq k \leq i$ o $ i<k<j$ la invariante es correcta


**Mantención:**

En la inesima iteración al ser un algoritmo que "divide y vencerás" hay dos posibles casos, la primera que $arreglo[j]>x$ y la segunda $arreglo[j]\geq x$
en la que el primer caso al iteración no cambia el arreglo y j se incrementa y en el segundo caso arreglo[i+1] se intercambia con arreglo[j] e i incrementa por lo que la función Partición es correcta



**3.3 Teorema correctitud función Quicksort**

Entra un arreglo de n elementos y sale un arreglo con la misma cantidad de elementos pero ordenados de menor a mayor.

**3.4 Prueba correctitud función Quicksort**

Para probar nuestro teorema deberemos utilizar la inductividad dado que la función quicksort es una función recursiva, en la cual debemos comprobar dos casos, el caso base y el paso inductivo en el cual tenemos una declaración que es P(n) para poder probar la inductividad:


P(n) se mantiene para todo $n\geq1$

**Caso base:**

Con cada arreglo de longitud = 1 ya está ordenado, por lo que quicksort retorna el arreglo de entrada, por lo que es correcto.

**Paso inductivo:**

en la iteración n-1 se crearán dos particiones a[n/2] en cada uno de los arreglos sé ira dividiendo hasta quedar con un elemento, por lo que este está ordenado, por lo que quicksort es correcto





# 4.- Tiempo de ejecución

**Teorema**

quicksort tiene una complejidad de $O(n^2)$ en el peor caso y $O(nlogn)$ en el mejor caso 

**Prueba teorema**

Para poder probar este teorema se va a utilizar el método de sustitución el cual consiste en tener una suposición de complejidad que tiene este y hacer una seria de cálculos con la ecuación de recurrencia del algoritmo.

Para probar de una forma más clara la complejidad de Quicksort vamos a asumir que la función partición siempre crea dos arreglos, uno de 0 elementos y el otro con n-1 elementos, por lo que queda la siguiente ecuación de recursión:

$T(n)=T(n-1) +T(0)+ O(n)$

Ahora con esta ecuación podemos empezar el método de sustitución en el cual vamos a tomar como suposición que en el peor caso quicksort tiene una complejidad de O(n^2) lo que da a los siguientes cálculos:
