¿Qué quiere decir que un problema NP-Hard? 

Lo que quiere decir que un problema es np-hard es un problema el cual está en el conjunto de np-completos pero no necesariamente en un problema np, ya que la dificultad de estos problemas con por lo menos igual de difíciles que los problemas np completos, pero estos puede que no sean siquiera verificables en un tiempo polinomial por lo que estos no necesariamente corresponde al conjunto np.

Un problema se puede saber si es np hard por medio de la reducción, puesto que si un problema verificado que es np-hard se puede reducir en nuestro problema se puede saber si es np hard.

¿Qué quiere decir que $P\neq NP$ ?

Esta conjetura quiere decir que los problemas p son distintos a los problemas np, ya que se sabe que los problemas p se pueden verificar y resolver en un tiempo polinomial sin importar el tamaño del problema, en cambio, los problemas np son problemas que se pueden verificar una solución en un tiempo de ejecución polinomial, pero estos no se pueden solucionar en un tiempo de ejecución polinomial, por lo que la conjetura se dice que es cierta, pero esta conjetura hay una pequeña posibilidad de que no sea así, sino, que p=np. Esto se dice porque tan solo basta que un problema considerado np se pueda resolver en tiempo polinomial y la conjetura $P\neq NP$ ya no seria validad, ya que ese problema np ya no es np sino p.

# 1. Problema 1 (Reducción de TSP a Camino Hamiltoniano)

Los problemas que estudiaremos son los problemas TSP (Traveling Salesman Problem), también conocido como problema del vendedor viajero, y el problema de Camino Hamiltoniano o Ciclo Hamiltoniano, el cual queremos reducir para probar que TSP es **NP-Hard**

## 1.1 Problema B: TSP (Traveling Salesman Problem)

El problema del vendedor viajero TSP se trata de un vendedor que quiere pasar por $n$ ciudades, y debe pasar por todas estas una única vez y volver al punto de partida.

Este problema tiene muchos algoritmos que tratan de solucionarlo en tiempo polinomial, pero no se ha tenido exito dado que no se logran obtener soluciones óptimas.

Un algoritmo simple que trata de resolver este problema es el de la fuerza bruta, el cual consiste en probar todas las posibles combinaciones del grafo en cuestión, la cual claramente no es óptimo. En este método de solución la cantidad de comparaciones realizadas está definido por la ecuación de $(n-1)!$ *(n-1 factorial)* siendo $n$ la cantidad de nodos que tiene el grafo.

Por ejemplo si el grafo de entrada tiene $5$ nodos, la ecuación se resolvería como $(5-1)*(4-1)*(3-1)(2-1)=24$, visto así parecen pocas comparaciones, pero si tan solo aumentamos el tamaño del grafo a $10$ nodos la cantidad de comparaciones se dispara a $362880$.

## 1.2 Problema A: Camino Hamiltoniano

El problema del camino hamiltonaiano consiste en una ruta o ciclo que pasa tan solo una vez por cada nodo del un grafo, el ciclo se genera porque este camino tiene que volver al punto de partida.

A los grafos que contienen un ciclo hamitoniano o camino hamitoniano se les denomina como grafo hamiltoniano y este problema se considera de la clase *NP-Hard*, ya que para poder encontrar este ciclo hamiltoniano se tiene que realizar una búsqueda completa y exhaustiva revisando todas las posibles combinaciones.

Un ejemplo de esto es cuando un grafo completo de más de dos nodos es un grafo hamiltoniano, esto es, porque un grafo completo como dice su definición es un grafo simple en el que cada nodo está conectado por un par de nodos únicos.

Este problema tiene diversas aplicaciones como por ejemplo diseño de circuitos eléctricos, mapeo de genomas, investigacion de operaciones, etc.

Una aplicación para ejemplificar es la ruta de un autobús escolar, ya que este tiene que recoger a todos los alumnos de un colegio en el cual se puede considerar que cada parada para recoger a un alumno es un nodo, el camino entre un nodo y el otro es el vértice y el autobús es el que quiere recorrer la ruta tan solo una vez.




## 1.3 Reducción de problema NP-Hard

En este caso se quiere comprobar cuál es la clase del problema **TSP** *(Traveling salesman Problem)* y para esto nos vamos a ayudar del problema explicado anteriormente, es decir, el problema del camino hamitoniano o ciclo hamiltoniano.

El problema del camino hamitoniano está verificado que es de la clase **NP-Hard**, por lo cual se buscará comprobar si el problema del camino hamiltoniano es reducible al problema TSP (Traveling Salesman Problem).

Para poder reducir el problema primero tenemos que saber cuáles son sus entradas y salidas:

**1. Ciclo hamintoniano**

**Entrada:** Un grafo $G=(V,E)$ siendo $V$ el conjunto de nodos del grafo y $E$ el conjunto de arcos del grafo.

**Salida:**Un ciclo que pasa por todos los nodos del grafo, pasando tan solo una vez por cada nodo y volviendo al punto de partida.

**2. TSP(taveling salesman problem)**

**Entrada:** Un grafo ponderado $G'=(V',E')$, sin pesos negativos, en la cual $V'$ es la cantidad de nodos del grafo y $E'$ la cantidad de arcos del grafo.

**Salida:** La combinación óptima de menor coste, la cual pasa por todos los nodos del grafo $G'$ y vuelve al nodo de partida.

Para reducir el problema del camino hamitoniano al problema de TSP para un grafo dado $G=(V,E)$, siendo $G$ un grafo completo, agregando aristas entre todos los pares de arcos que no estaban conectados en $G$ tenemos que ver el grafo nuevo $G'=(V',E')$ donde $V'=V$ y $E'={(u,v)}$ donde cada $u,v$ pertenece a $V'$.

## 1.4 Algoritmos que resuelven el problema B (TSP)

El problema del TSP tiene muchos algoritmos que tratan de resolverlo, pero ninguno de estos ha tenido éxito de resolverlo en tiempo polinomial, el más sencillo de estos es el método de la fuerza bruta, el cual consiste que por medio de la búsqueda exhaustiva y completa busca todas las posibles combinaciones que tiene el grafo de $n$ nodos, esto significa que si tenemos un grafo con $5$ nodos la cantidad de combinaciones posibles que tiene este es de $24$, pero si nos vamos a un grafo con $10$ nodos el resultado se dispara a $362880$ posibles combinaciones, esto se debe, ya que la cantidad de combinaciones que tiene el algoritmo de fuerza es de $(n-1)!$, siendo n el número de nodos que tenga el grafo, dada esta ecuación el tiempo de ejecución de esta no se queda atras, ya que esta es $O((n-1)!)$ siendo la misma que la cantidad de comparaciones.

Otro algoritmo que trata de resolver este problema es por método de la programación dinámica que consiste en un conjunto de $n$ nodos $\{ 1,2,3 \ldots n \}$ el cual asignamos el nodo $1$ como inicial y final.

Para cada nodo $L$ distinto de $1$ buscamos la ruta de costo mínimo con el nodo $1$ como el nodo de partida y $L$ el nodo final en la cual todos los nodos salen solo una vez.

El costo del camino será $i$, y el coste del ciclo corresponde a $i$+$\text{dist}(i,1)$ donde dist corresponde la distancia entre el nodo $L$ y el nodo $1$, luego para terminar devolvemos el mínimo de todos los valores $cost(i)+dist(i,1)$.

Ahora se tiene que definir una función recursiva, la cual será $C(S,i)$ como el costo mínimo del camino que visita cada nodo una vez del conjunto $S$, comenzando en $1$ y terminando en $i$. 

Ahora comenzamos todos los subconjuntos de tamaño $2$ y calculamos $C(S,i)$, luego calculamos todos los subconjuntos de tamaño $3$ con $C(S,i)$ y así sucesivamente donde el nodo $1$ tiene que estar presente en todos los subconjuntos.

Para un conjunto de tamaño $n$ consideramos $n-1$ subconjuntos, los cuales cada uno tiene un tamaño de $n-1$, de este modo los subconjuntos no tienen $n$-esimos en ellos.

Usando la recurrencia anterior podemos usar la programación dinámica en el cual hay un máximo de $O(n*2^n)$ subproblemas, y cada uno requiere un tiempo lineal para resolverse, por lo cual el tiempo de ejecución de este algoritmo es de $O(n^2*2^n)$.

La complejidad de esta comparado a la de fuerza bruta es mucho menor, pero esto no significa que sea eficiente, ya que este sigue teniendo un tiempo exponencial, esta solución nos permite un tiempo de ejecución menor en casos de que el grafo tenga una mayor cantidad de nodos que la cantidad de nodos de la fuerza bruta, pero sigue sin ser una solución.

# 2. Problema 2 (Reducción de Independent Set a Clique)

Ahora queremos probar que el problema Clique es **NP-Hard**, por lo que trataremos de reducir el problema Independent Set a Clique.

## 2.1 Problema A: Independent Set

En teoría de grafos, un conjunto independiente o conjunto estable, es un conjunto de nodos en un grafo de tal forma que ninguno de estos es adyacente a otro, es decir un conjunto $V$ de nodos tal que $\forall (u,v) \in V, \nexists \ E(u,v)$ *(para cada par de nodos en $V$, no debe existir ningun arco entre ellos)*.

Sabemos que este problema es **NP-Hard**

## 2.2 Problema B: Clique

En teoría de grafos, un Clique en un grafo no dirigido $G=(V,E)$ es un conjunto de nodos $C \subseteq V$ tal que todo par de vértices dentro del conjunto son adyacentes entre sí, es decir, un Clique es un sub-grafo en el cual cada nodo está conectado a los demás nodos del mismo sub-grafo

## 2.3 Reducción de problema NP-Hard

Primero definiremos las entradas y salidas de cada problema.

### Clique

**Entrada:** Un grafo $G=(V,E)$ y un entero $K$ que corresponde al tamaño del Clique buscado. 

**Salida:** Un conjunto de nodos que cumplan con ser Clique de tamaño $\geq K$.

### Independent Set

**Entrada:** Un grafo $G'=(V',E')$ y un entero $K'$.

**Salida:** Un conjunto de nodos que cumplan con ser un Conjunto Independiente de tamaño $\geq K'$.

### **Reducción**

Para entender esta reducción, es importante darse cuenta que si un grafo tiene un Conjunto Independiente de tamaño $K$, entonces el complemento de ese grafo (o grafo inverso) posee un Clique de tamaño $K$.

Formalmente, podemos decir que dado el grafo $G=(V,E)$ donde existe un conjunto independiente de tamaño $\geq K$, podemos producir el complemento $\overline{G}=(V,\overline{E})$ donde existirá un Clique de tamaño $\geq K$

Entonces podemos decir que el problema Independent Set puede ser reducido a Clique mediante los siguientes pasos:

1. Calculamos el complemento $\overline{G}$ de $G$ 
2. Encontramos un Clique para el complemento $\overline{G}$
3. El Clique de $\overline{G}$ corresponde a un conjunto independiente de $G$

Por lo que queda demostrado que **Clique es NP-Hard**.

## 2.4 Algoritmos que resuelven el problema B (Clique)

Un algoritmo simple (pero ineficiente) que resuelva el problema Clique puede ser utilizando la fuerza bruta.

1.   Generamos todos los subgrafos posibles de tamaño $K$.
2.   Verificamos si cada subgrafo corresponde a un Clique válido. 

Este algoritmo tiene tiempo de ejecución $O(V^K K^2)$ donde $V$ es la cantidad de nodos del grafo y $K$ es el tamaño del Clique buscado.

Esto se puede demostrar, ya que existen $\binom{V}{K}$ conjuntos de $K$, para verificar si estos conjuntos son un Clique válido debemos verificar entonces $\binom{K}{2}$ pares de nodos para verificar si hay arcos entre ellos, es decir, debemos realizar $\binom{V}{K} \binom{K}{2}$ comparaciones.
