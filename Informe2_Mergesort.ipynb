# 2.- Merge and Mergesort
**2.1 Descripción función Mergesort**

El algoritmo mergesort es un algoritmo de ordenamiento que se basa en el método divide y vencerás que consiste en ir dividiendo el problema cada vez problemas más pequeños hasta tener la solución de todos estos "subproblemas" y poder resolver el problema mayor.

Lo que hace este algoritmo es ir dividiendo el arreglo, una y otra vez hasta que estos tienen un tamaño de 1, luego estos entran en la función merge.

**Pasos:**

-comprueba si el arreglo de n-1 elementos, su "len" o tamaño sea menor a 2, si es este el caso este retorna el arreglo, la cual tiene un elemento y ya está ordenada.

-divide el arreglo en dos mitades, y estas entran nuevamente al mergesort para ser dividida hasta que este cumpla con la condición de len(lista)<2.

-así se le asigna un arreglo "izquierda" y "derecha" los cuales son pasados a la función merge

**2.2 Descripción función merge**


En esta función se toman los arreglos de n elementos y ve cuál de los dos tiene el menor elemento (lista1[i] < lista2[j]), al comprobarlo el menor elemento se mete a un arreglo auxiliar, luego se van introduciendo en este arreglo auxiliar para ser retornado, en esta función van a ir entrando los mismos arreglos que esta función ya ordeno para finalmente retornar un arreglo con la misma cantidad de elementos que el arreglo de entrada, pero ordenada de menor a mayor.

**Pasos**

-entran los arreglos A y B

-se recorre con un while el len de A y B

-se compara si la i-1 posición de arreglo A[i-1] es menor al arreglo B[j-1], si es así se agrega al nuevo arreglo la el i-1 elemento del arreglo A[i-1].

-si esta comparación es falsa, se ingresa al nuevo arreglo el elemento de B en la j-1 posición(B[j-1])

**2.3 Implementación función Merge**

A continuación la implementación de la función Merge

```python

def merge(lista1, lista2, verbose):

    i, j = 0, 0
    result = [] 
    global cont
    
    if (verbose==True):print("derecha= ",lista1)
    if (verbose==True):print("izquierda= ",lista2)
   #ordenar en arreglo result de menor a mayor
    while(i < len(lista1) and j < len(lista2)):
        
        if (lista1[i] < lista2[j]):
            result.append(lista1[i])
            
            i += 1
            cont+=1
            
        else:
            result.append(lista2[j])
            
            j += 1
            cont+=1
            
 
   #agregar resultados a la lista 
    result += lista1[i:] + lista2[j:]
    
    if (verbose==True):print("Merge= ",result)
    
   
    cont+=1
    # Retornamos el resultados
    return result
```

**3.4 teorema 2 correctitud mergesort**

El mergesort ordena en un arreglo $[1,54,2...,n]$ y entrega un arreglo con los mismos elementos pero ordenados de menor a mayor $[1,2,54....,n]$


**3.5 correctitud mergesort**





Para probar la correctitud del mergesort, al ser una función recursiva tenemos que utilizar la inducción, en la cual tenemos que tener una declaración, que será $P(n)$, en la cual se tiene que probar el **caso base** y un **paso inductivo**



**paso base:**

En el caso de que $n=1$ tenemos que $P(1)$ mergesort retorna el elemento, ya que este es solo un elemento, por lo que se cumple $P(n)$

**caso inductivo**

En este caso hay que probar que P(m) es cierta para m<n

En el caso de n elementos mergesort divide el arreglo en dos subarreglos que toman el tamaño de $m=n/2$ y $m=n/2$ respectivamente, como el tamaño de m es siempre menor a n sin importar cuantas veces se divida, asumimos que en el paso inductivo los arreglos son ordenados correctamente, como el merge retorna los arreglos combinados, este arreglo terminan con los mismos elementos que el arreglo de entrada, pero ordenados de menor a mayor, por lo tanto, mergesort es correcto.

# 4.- Tiempo de ejecución
**4.1 Teorema**

El algoritmo mergesort tiene una complejidad temporal de $O(nlog(n))$

**4.2 Prueba teorema**

Dado que el algoritmo del Mergesort es un algoritmo que está basado principalmente en el método divide y vencerás, este tiende a dividir constantemente en arreglo entregado, hasta dejarlo con 1 elemento.
Como Mergesort es un algoritmo recursivo, se tiene que utilizar una ecuación recurrente.

El algoritmo mergesort siempre está dividiendo el arreglo en dos, por lo cual el siguiente subproblema siempre será un 1/2 del problema anterior, y al dividir el arreglo en dos el tiempo siempre es el mismo, por lo que la complejidad de este que queda $O(1)$

La complejidad del merge al ser lineal queda como O(n) lo cual resulta la siguiente ecuación recurrente



$T(n) = \begin{cases}
  O(1)  & n \text{ is small} \\
  2 T(n/2) + O(n) & \text{ otherwise}
\end{cases}$

Teniendo esta ecuación tenemos una manera de probar que nuestro teorema.

**4.3 Árbol de recurrencia mergesort**

Para probar el teorema con el árbol de recurrencia primero hay que modificar la ecuación de recurrencia dejando $O(1)$ como una constante $c$ y la complejidad $O(n)$ como $c*n$ dejando la ecuación:


$T(n) = \begin{cases}
  c  & n \text{ is small} \\
  2 T(n/2) + c \cdot n & \text{ otherwise}
\end{cases}$

ahora se puede crear el árbol de recurrencia dividiendo cada caso en $T(n)$ recursivamente

Teniendo esto podemos probarlo definiendo n con un valor, por ejemplo n=8, el resultado de $log(n)+1=4$ que son los niveles que tendría el arbol, ya que al generar 8 ramas esta tiene una profundidad de 4 niveles, entonces el tiempo total de mergesort es de $cn(log(n)+1)$, pero por la notación de la complejidad se va la constante y el +1 quedando $O(n(log(n))$
